5. В прошлом месяце Иван приобрел акции компании, занимающейся разработкой ПО.

Было куплено 3500 акций. Иван заплатил 1000.00 рублей за акцию, кроме этого, ему пришлось заплатить комиссию банку, которая составила 2.5% от суммы стоимости акций.

Через месяц Иван продал все акции. Стоимость продажи составила 1180 рублей за акцию и
комиссия банка также составила 2.5% от суммы, полученной за акции.

Напишите программу, которая рассчитывает: сумму денег, уплаченную за акции и сумму комиссии, которую получил банк при покупке; сумму денег за которую Иван продал акции и сумму комиссии, уплаченную при продаже.

А также рассчитывает сумму дохода или убытка Ивана от операций с акциями.


```python
print("ЗАДАНИЕ 5")
# Входные данные
num_shares = 3500
buy_price_per_share = 1000.00
sell_price_per_share = 1180.00
commission_rate = 2.5 / 100  # 2.5% в виде десятичной дроби

# 1. Покупка акций
purchase_amount = num_shares * buy_price_per_share
purchase_commission = purchase_amount * commission_rate
total_purchase_cost = purchase_amount + purchase_commission

# 2. Продажа акций
sale_amount = num_shares * sell_price_per_share
sale_commission = sale_amount * commission_rate
total_sale_income = sale_amount - sale_commission

# 3. Прибыль или убыток
profit_or_loss = total_sale_income - total_purchase_cost

# Вывод результатов
print(f"Сумма денег, уплаченная за акции: {purchase_amount} рублей")
print(f"Комиссия банка при покупке: {purchase_commission} рублей")
print(f"Сумма денег, полученная от продажи акций: {sale_amount} рублей")
print(f"Комиссия банка при продаже: {sale_commission} рублей")
print(f"Итоговая прибыль/убыток: {profit_or_loss} рублей")
```

7. Площадь прямоугольника- это произведение его длины на его ширину. Напишите программу, которая запрашивает длину и ширину двух прямоугольников. Программа должна выводить пользователю значения площадей каждого прямоугольника, а также сообщение о том, площадь какого прямоугольника больше, либо сообщать, что они одинаковы.

```python
length_first = float(input("Введите длину первого прямоугольника: "))
width_first = float(input("Введите ширину первого прямоугольника: "))

length_second = float(input("Введите длину второго прямоугольника: "))
width_second = float(input("Введите ширину второго прямоугольника: "))

area1 = length_first * width_first
area2 = length_second * width_second

print(f"Площадь первого прямоугольника: {area1:.2f}")
print(f"Площадь второго прямоугольника: {area2:.2f}")

if area1 > area2:
  print("Площадь первого прямоугольника больше.")
elif area1 < area2:
  print("Площадь второго прямоугольника больше.")
else:
  print("Площади прямоугольников одинаковы.")
```

8. Дата 10 июня 1960 года является особенной, потому что если ее записать в формате 10.06.60, то произведение дня и месяца равняется году. Разработайте программу, которая просит пользователя ввести месяц (в числовой форме), день и год (две последние цифры).  Затем программа должна проверить корректность введенных данных и, если все верно, определить, равняется ли произведение дня и месяца году. Если это так, то она должна вывести сообщение, что введенная дата является волшебной. В противном случае она должна вывести сообщение, что дата не является волшебной.

```python
def is_magic_date(day, month, year):
  """Проверяет, является ли дата "волшебной", т.е. произведение дня и месяца равно году."""
  if not 1 <= month <= 12 or not 1 <= day <= 31:
    return False

  # Проверка количества дней в месяце
  if month == 2 and day > 29:
    return False
  if month in [4, 6, 9, 11] and day > 30:
    return False
  
  return day * month == int(str(year)[2:])


# Получение данных от пользователя
day = int(input("Введите день (число): "))
month = int(input("Введите месяц (число): "))
year = int(input("Введите год (последние две цифры): "))

# Проверка и вывод результата
if is_magic_date(day, month, year):
  print("Введенная дата является волшебной!")
else:
  print("Введенная дата не является волшебной.")
```

* для карманов с 1 по 10 карманы с нечетным номером имеют красный цвет, карманы с четным номером - черный;
* для карманов с 11 по 18 карманы с нечетным номером имеют черный цвет, карманы с четным номером - красный;
* для карманов с 19 по 28 карманы с нечетным номером имеют красный цвет, карманы с четным номером - черный; 
* для карманов с 29 по 36 карманы с нечетным номером имеют черный цвет, карманы с четным номером - красный. 

Напишите программу, которая просит пользователя ввести номер кармана и показывает, является ли этот карман зеленым, красным или черным. Программа должна вывести сообщение об ошибке, если пользователь вводит число, которое лежит вне диапазона от 0 до 36.


```python
number = int(input("Введите номер кармана (от 0 до 36): "))

if number == 0:
  color = "зеленый"
elif 1 <= number <= 10 or 19 <= number <= 28:
  color = "красный" if number % 2 != 0 else "черный"
elif 11 <= number <= 18 or 29 <= number <= 36:
  color = "черный" if number % 2 != 0 else "красный"
else:
  color = "Ошибка: номер кармана должен быть от 0 до 36"

print(f"Карман {number} - {color}")
```

```python
total_coins = 0

while True:
    print("Введите количество монет каждого достоинства:")
    coins_5 = int(input("Монеты 5 копеек: "))
    coins_10 = int(input("Монеты 10 копеек: "))
    coins_50 = int(input("Монеты 50 копеек: "))

    total_coins = coins_5 * 5 + coins_10 * 10 + coins_50 * 50

    if total_coins == 100:
        print("Поздравляем! Вы собрали 1 рубль!")
        break
    elif total_coins > 100:
        print("К сожалению, вы собрали больше 1 рубля.")
    else:
        print("К сожалению, вы собрали меньше 1 рубля.")
```

12. Бег на беговой дорожке позволяет сжигать 5 калорий в минуту. Напишите программу, которая применяет цикл для вывода количества калорий, сожженных после 10, 15, 20, 25 и 30 минут бега.

```python
def burn_kk(mins):
    count = 0
    for _ in range(mins):
        count += 5
    return count


print(f"Сожжено каллорий за 10 минут: {burn_kk(10)}")
print(f"Сожжено каллорий за 15 минут: {burn_kk(15)}")
print(f"Сожжено каллорий за 20 минут: {burn_kk(20)}")
print(f"Сожжено каллорий за 25 минут: {burn_kk(25)}")
print(f"Сожжено каллорий за 30 минут: {burn_kk(30)}")
```

14. Напишите программу с циклом, которая просит пользователя ввести ряд положительных чисел. Пользователь должен ввести отрицательное число в качестве сигнала конца числового ряда. После того как все положительные числа будут введены, программа должна вывести их сумму.

```python
total = 0
number = 0

while number >= 0:
  number = int(input(
      "Введите положительное число (введите отрицательное число для завершения): "))
  if number >= 0:
    total += number

print("Сумма чисел:", total)
```

15. В заданном университете обучение студента-заочника составляет 90 000 рублей в год. Было объявлено, что плата за обучение будет повышаться на 5% каждый год в течение следующих 5 лет. Напишите программу с циклом, который выводит плановую сумму за обучение в семестр в настоящий момент и в течение 5 следующих лет.

```python
price_year = 90_000 
percent = 0.05

for year in range(1, 7):
    semester_fee = price_year / 2  # Стоимость за семестр
    print(f"Год {year}: {semester_fee:.2f} рублей")
    price_year *= (1 + percent)  # Увеличение стоимости на 5%
```
